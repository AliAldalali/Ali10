```python
grades = []
is_failed = False

print("الرجاء إدخال درجات المواد الخمس:")

for i in range(1, 6):
    while True:
        try:
            grade = float(input(f"أدخل درجة المادة رقم {i}: "))
            if 0 <= grade <= 100:
                grades.append(grade)
                if grade < 50:
                    is_failed = True
                break
            else:
                print("الدرجة يجب أن تكون بين 0 و 100. الرجاء إعادة المحاولة.")
        except ValueError:
            print("إدخال غير صحيح. الرجاء إدخال رقم.")

if not grades: # Should not happen if loop runs successfully, but good practice
    print("لم يتم إدخال أي درجات.")
else:
    total_grades = sum(grades)
    average = total_grades / len(grades)

    print(f"\nمعدل الطالب: {average:.2f}")

    if is_failed:
        print("التقدير: راسب ")
    else:
        if 90 <= average <= 100:
            print("التقدير: ممتاز")
        elif 80 <= average < 90:
            print("التقدير: جيد جداً")
        elif 70 <= average < 80:
            print("التقدير: جيد")
        elif 60 <= average < 70:
            print("التقدير: مقبول")
        else: # This covers 50-59 if no individual grade was below 50
            print("التقدير: ضعيف")
```
